```python
def draw_graph(t_acc, v_acc, label, ax=None, title='Learning Curve', legend_loc='lower right'):
    # Accuracy Vs Epocs
    t_acc = np.array(t_acc)
    v_acc = np.array(v_acc)
    if ax is None:
        fig, ax = plt.subplots(figsize=(10, 8))
        ax.set_title(title, fontsize=20)
        ax.set_ylabel('Accuracy')
        ax.set_xlabel('Epochs')
        epochs = len(t_acc)
        ax.set_xlim([-5, epochs*1.05])
        
        ax.set_yscale('log')

    
    min_accuracy = min(np.min(t_acc), np.min(v_acc))
    if ax.get_ylim()[0] > 10:        
        min_accuracy = min(min_accuracy, ax.get_ylim()[0])
    ax.set_ylim([min_accuracy, 105])

    plt_x = list(range(1, len(t_acc) + 1, 1))
    # print(len(plt_x), len(t_acc), len(v_acc))
    ax.plot(plt_x, t_acc, label=f"{label} train")
    ax.plot(plt_x, v_acc, label=f"{label} val")
    ax.legend(loc=legend_loc)

    return ax
```
